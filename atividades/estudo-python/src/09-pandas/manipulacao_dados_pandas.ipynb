## Criando um DataFrame

Um novo DataFrame pode ser criado a partir de um dicionário, onde cada chave possui uma linha de elementos de igual tamanho.
    As chaves representam as colunas e cada um dos valores de sua lista representa o valor da linha correspondente para aquela coluna.

````python
personagens_df = pd.DataFrame({
    'nome': ['José da Silva', 'Yoda', 'Luke Skywalker'] ,
    'idade': [50, 100, 20] ,
    'peso': [70.3, 15.8, 60.5] ,
    'eh jedi': [False, True, True] # nomes nas colunas pode ter espaço
})
````

````python
personagens_df
````

````python
personagens_df.info()
````

## Renomeando as colunas de um DataFrame

O atributo DataFrame.columns retorna uma lista com os nomes de todas as colunas do data frame.

````python
personagens_df.columns
````

````python
# converte para uma lista 
list(personagens_df.columns)
````

===> Para renomear colunas do data frame, utilizamos o método DataFrame.rename, que retorna uma cópia do data frame com as colunas renomeadas:

````python
personagens_df.rename(columns={
    'nome': 'Nome Completo',
    'idade': 'Idade',
    'peso': 'Peso',
    'eh jedi': 'Eh Jedi'
})
````

````python
personagens_df
````

### para alterar os valores originais usamos inplace = True

````python
personagens_df.rename(columns={
    'nome': 'Nome Completo',
    'idade': 'Idade',
    'peso': 'Peso',
    'eh jedi': 'Eh Jedi'
}, inplace = True)
````

````python
personagens_df
````

### renomear todas as colunas de uma única vez

````python
lista = [ x.upper() for x in list(personagens_df.columns) ]
personagens_df.columns = lista
personagens_df
````

Caso faça:
```bash
    produto = data['PRODUTO'] # produto é apenas referencia
```

Temos que produto é apenas uma referencia , ou seja caso não queiramos alterar diretamente os valores de nosso data frame devemos fazer uma cópia do mesmo para uma nova variáveil usando o comando copy() não perdendo valores e preservando o data frame original.

## Criando novas colunas
Para criar uma nova coluna no data frame , atribuímos uma lista ou Series ou uma constante a uma nova chave no 'data frame'.   
A quantidade de valores da lista precisa ser igual ao número de linhas/registros do data frame

````python
# criando uma nova coluna no data frame , como iremos fazer isso apenas para facilitar 
# irei criar uma cópia de nosso DataFrame original
data_copia = data.copy()
data_copia
````

### Criando uma coluna a partir de um valor constante

````python
data['NOVA COLUNA'] = 'DEFAULT'
data.head()
````

Ao criar o comando eu digitei duas vezes e com isso temos duas colunas novas, uma que se chama 'Nova Coluna' e outra que se chama 'NOVA COLUNA', com isso temos um produto semântico , que devemos resolver apagando uma das colunas novas usando o comando drop()

````python
data = data.drop(columns=['Nova Coluna'])
data
````

### Atribuindo intervalo de valores para uma nova coluna

````python
data['INTERVALO DE VALORES'] = range(data.shape[0])
data
````

### Atribuindo uma Series a uma nova coluna no dataframe

````python
# verificar se os valores são do mesmo tipo
print(data['PREÇO MÉDIO DISTRIBUIÇÃO'].dtypes)

# realizar a conversão para o tipo adequado
data['PREÇO MÉDIO DISTRIBUIÇÃO'] = pd.to_numeric(data['PREÇO MÉDIO DISTRIBUIÇÃO'], errors='coerce')

data['PREÇO MÉDIO DISTRIBUIÇÃO EM DÓLAR'] = data['PREÇO MÉDIO DISTRIBUIÇÃO'] * 5.10
data
````
